```python
def creating_user_item_df(temp):

  temp['description'] = [str(i[2:-2]) for i in temp['description']]

  # reset index
  temp = temp.reset_index()
  temp.drop(['index'], axis = 1, inplace = True)

  #Removing empty descriptions
  temp = temp[temp['description'] != '[]']

  # Removed items from Unknown brands (191 entries)
  temp = temp[temp['brand'] != 'Unknown']

  temp_user = temp[['reviewerID', 'reviewerName', 'reviewText', 'summary','asin']]
  user_ltr = temp[['reviewerID', 'asin', 'overall']]
  temp_item = temp[['description', 'title', 'brand', 'price', 'asin']]

  #creating user features by groupBy

  temp_x = temp_user.groupby('reviewerID')

  user = pd.DataFrame(columns=['reviewerID', 'reviewerName', 'reviewText', 'summary','asin'])

  for name, group in temp_x:
    user = user.append({'reviewerID': group['reviewerID'].tolist()[0],
                'reviewerName': group['reviewerName'].tolist()[0],
                'reviewText': group['reviewText'].tolist(),
                'summary': group['summary'].tolist(),
                'asin': group['asin'].unique().tolist()}, ignore_index = True)
    

  #creating item features by groupBy

  temp_x = temp_item.groupby('asin')

  item = pd.DataFrame(columns=['asin', 'description', 'title', 'brand', 'price'])

  for name, group in temp_x:
    item = item.append({'asin': group['asin'].tolist()[0],
                'description': group['description'].tolist()[0],
                'title': group['title'].tolist()[0],
                'brand': group['brand'].tolist()[0]}, ignore_index = True)
  return user, item, user_ltr
```

```python
from sent2vec.vectorizer import Vectorizer
def sentence_embeddding(df, review_list, summary_list):
  
  user_review_vec = []
  user_summary_vec = []

  vectorizer = Vectorizer()
  for i in range(1,269):
    vectorizer = Vectorizer()
    j = 100*i
    k = j-100
    vectorizer.run(review_list[k:j])
    user_review_vec.extend(vectorizer.vectors)

    vectorizer.run(summary_list[k:j])
    user_summary_vec.extend(vectorizer.vectors)

  vectorizer.run(review_list[26800:])
  user_review_vec.extend(vectorizer.vectors)

  
  vectorizer.run(review_list[26800:])
  user_summary_vec.extend(vectorizer.vectors)

  
  df['user_review_embs'] = user_review_vec
  df['user_summary_embs'] = user_summary_vec

  df.to_pickle('user_emb.pickle')
```

```python
from sent2vec.vectorizer import Vectorizer
def sentence_embeddding_item(df):

  des_list = []
  title_list = []
  brand_list = []

  for i in range(len(df)):
    des_list.append(str(df.description)[2:514])


  item_des_vec = []
  item_title_vec = []
  item_brand_vec = []

  vectorizer = Vectorizer()
  vectorizer.run(des_list)
  item_des_vec.extend(vectorizer.vectors)
  df['item_des_embs'] = item_des_vec
  
  vectorizer = Vectorizer()
  vectorizer.run(df['title'].tolist())
  item_title_vec.extend(vectorizer.vectors)
  df['item_title_embs'] = item_title_vec

  vectorizer = Vectorizer()
  vectorizer.run(df['brand'].tolist())
  item_brand_vec.extend(vectorizer.vectors)
  df['item_brand_embs'] = item_brand_vec

  df.to_pickle('item_emb.pickle')
```

```python
#Evaluating predictions

from sklearn.metrics import mean_squared_error
from math import sqrt
def RMSE(predicted, actual):
    predicted = predicted[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    
    return sqrt(mean_squared_error(predicted, actual))
```

```python
def recommend_top_n_user_based(user_id, top_n=5):
  #dictionary reverse for old user ids
  for old, new in new_user_ids.items():
    if user_id == old:
      print("Top ", top_n, "products for user_id", user_id)
      #print("Actual user id", old)
  
  item_rated = list(new_df['asin'].loc[new_df['reviewerID'] == user_id])
  temp = user_predictions_df.loc[user_predictions_df['reviewerID'] == user_id].copy()
  temp.drop(user_predictions_df[item_rated], axis=1, inplace=True) #remove already rated products
  final = temp.iloc[:, 1:].sort_values(by=temp.index[0], axis=1, ascending=False) #descending order of similarity
  top_recommend = final.iloc[:, :top_n].to_dict(orient='records')

  return top_recommend
```

```python
from tqdm import tqdm
def get_user_features(df):
    res = list()
    for i, v in tqdm(df.groupby('userid')):
        res.append(
            (
                i,
                len(v['itemid']),
                (v['rating'] == 5).sum(),
                (v['rating'] == 4).sum(),
                (v['rating'] == 3).sum(),
                (v['rating'] == 2).sum(),
                (v['rating'] == 1).sum()
            )
        )
    
    res = pd.DataFrame(
        res,
        columns=[
            'userid', 'item_count', '5_star_ratings_gave', '4_star_ratings_gave',
            '3_star_ratings_gave', '2_star_ratings_gave', '1_star_ratings_gave'])
    return res
```

```python
def get_item_features(df):
    res = list()
    for i, v in tqdm(df.groupby('itemid')):
        res.append(
            (
                i,
                len(v['userid']),
                (v['rating'] == 5).sum(),
                (v['rating'] == 4).sum(),
                (v['rating'] == 3).sum(),
                (v['rating'] == 2).sum(),
                (v['rating'] == 1).sum()
            )
        )
    
    res = pd.DataFrame(
        res,
        columns=[
            'itemid', 'user_count', '1_star_ratings_recieved', '2_star_ratings_recieved',
            '3_star_ratings_recieved', '4_star_ratings_recieved', '5_star_ratings_recieved'])
    return res
```

```python
def predict_at_k(data, model, k):
    user_ids = list()
    product_ids = list()
    ranks = list()
    
    for userId, df in data.groupby('userid'):
        
        pred = model.predict(df.loc[userId])
        productId = np.array(df.reset_index()['itemid'])
        topK_index = np.argsort(pred)[::-1][:k]
        product_ids.extend(list(productId[topK_index]))
        user_ids.extend([userId]*len(topK_index))
        ranks.extend(list(range(1, len(topK_index)+1)))

    results = pd.DataFrame({'userid': user_ids, 'itemid': product_ids, 'rank': ranks})
    
    return results

predicted = predict_at_k(grouped_xte, model, 10)
```
